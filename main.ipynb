```python
import numpy as np
import pandas as pd
import yfinance as yf
import sklearn
import pandas_ta as ta
```

```python


# Set the ticker as 'EURUSD=X'
forex_data = yf.download('EURUSD=X', period='max',interval='1d')
# Display the last five rows
data = pd.DataFrame(forex_data)
data.tail(5)
```

```python
data["rsi"] = ta.rsi(data["Close"])
data["sma"] = ta.sma(data["Close"],lenght=14)
data["range"] = data['High'] - data['Low']
data.dropna()
data = data.drop(["Volume","Adj Close","Open", "Close"],axis=1)

data_model = data.head(-1)
today_data = data.tail(1)

today_data
```

```python
data_model["signal"] = 0  # Initialize with 0

# Condition 1: If condition 1 happens, set the value to 1 (Bull)
condition1 = (data_model['High'] > data_model['High'].shift(1)) & (data_model['Low'] > data_model['Low'].shift(1)) & ((data_model['High'] - data_model['High'].shift(1)) >= 0.0040)
data_model.loc[condition1, "signal"] = 1

# Condition 2: If condition 2 happens, set the value to 2 (Bear)
#condition2 = (data_model['Low'] < data_model['Low'].shift(1)) & (data_model['High'] < data_model['High'].shift(1)) & ((data_model['Low'].shift(1) - data_model['High']) >= 0.0010)
#data_model.loc[condition2, "signal"] = 2


data_model.dropna()
data_model=data_model.iloc[20:]
data_model.tail(50)
```

```python
true_count = data_model["signal"].sum()
print(f"Number of True values: {true_count}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
clf = RandomForestClassifier()

```

```python
X = data_model.drop(['signal'], axis=1)
y = data_model['signal']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(today_data)
```
